# Analysis

Convert nan values in `conservation_status` column to No Intervention

```python
species.fillna('No Intervention', inplace=True)
species.groupby("conservation_status").size()
```

## Distribution of conservation status for species

```python
conservation_category = species[species.conservation_status != 'No Intervention']\
    .groupby(['conservation_status','category'])['scientific_name']\
    .nunique()\
    .reset_index(name='species_count')
conservation_category
```

```python
pivot_data = conservation_category.pivot_table(index='conservation_status', 
                                              columns='category', 
                                              values='species_count', 
                                              aggfunc='sum', 
                                              fill_value=0)
status_order = ['Endangered', 'Threatened', 'Species of Concern', 'In Recovery']
pivot_data = pivot_data.loc[status_order]
pivot_data
```

```python
pivot_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Conservation Status')
plt.ylabel('Number of Species')
plt.title('Distribution of species by conservation status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## In conservation

```python
species['is_protected'] = species['conservation_status'] != 'No Intervention'
species.head()
```

```python
protected_category = species.groupby(['category','is_protected'])\
    .scientific_name.nunique()\
    .reset_index()\
    .pivot_table(index='category',
                columns='is_protected',
                values='scientific_name')\
    .reset_index()
protected_category.columns = ['category', 'not_protected', 'protected']
protected_category
```

Mammals are most likely to be protected (17%), followed by birds (15%). In contrast, only 1% of plants are under protection.

```python
protected_category['percent_protected'] = protected_category.protected/(protected_category.not_protected+protected_category.protected)*100
protected_category
```

## Statistical significance

Test the difference between mammal and bird
||protected|not protected|
|-|-|-|
|Mammal|?|?|
|Bird|?|?|

p-valus is 0.69 > 0.05, which means that there doesn't seem to be any significant relationship between mammal and bird.

```python
contingency1 = [[30, 146],
              [75, 413]]
chi2_contingency(contingency1)
```

Test the difference between mammal and reptile.
p-value is 0.04 < 0.05, which means that the difference between reptile and mammal is statistically significant. Mammals are shown to have a statistically significant higher rate of needed protection compared with Reptiles. 

```python
contingency2 = [[30, 146],
              [5, 73]]
chi2_contingency(contingency2)
```

## Species in parks

```python
species['common_names_split'] = species.common_names.str.replace(',','').str.split()
species['common_names_split'] = species['common_names_split'].apply(lambda x: list(set(x)))

# Display the resulting DataFrame
species[['common_names', 'common_names_split']]
```

Bat is the most prevalent animal

```python
mammal_counts = species[species.category=='Mammal'].common_names_split.explode().value_counts()
mammal_counts
```

```python
species['is_bat'] = species.common_names.str.contains(r'\bBat\b',regex=True)
species[species.is_bat]
```

```python
bat_observations = observations.merge(species[species.is_bat])
bat_observations
```

```python
bat_observations.groupby('park_name').observations.sum().reset_index()
```

```python
obs_by_park = bat_observations.groupby(['park_name', 'is_protected']).observations.sum().reset_index()
obs_by_park
```

Yellowstone and Bryce National Parks seem to be doing a great job with their bat populations since there are more sightings of protected bats compared to non-protected species.

```python
plt.figure(figsize=(16, 4))
sns.barplot(x=obs_by_park.park_name, y= obs_by_park.observations, hue=obs_by_park.is_protected)
plt.xlabel('National Parks')
plt.ylabel('Number of Observations')
plt.title('Observations of Bats per Week')
plt.show()
```
